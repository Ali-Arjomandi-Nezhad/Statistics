```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
```

Prior:

```python
a=3 
b=3
rv = beta(a, b)
```

```python
x=np.linspace(0, 1, 101)
Prior=rv.pdf(x)
plt.plot(x,Prior, color='red')
```

Observed Data:

```python
obsrv=np.array([0 , 0 , 0 , 1 , 1 , 0, 1 , 1 , 1 , 1])
```

Posterior:

```python
Posterior=np.array([])
for i in Prior:
    u=i
    for j in obsrv:
        u=u*pow(i,j)*pow(1-i, 1-j)    
    Posterior= np.append( Posterior , u  ) 
    
Posterior=Posterior/(sum(Posterior) *0.01)   
```

```python
plt.plot( x , Posterior )
```
