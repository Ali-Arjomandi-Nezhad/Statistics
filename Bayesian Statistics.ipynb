In this example we want to model the distribution of the probability that a coin gives head (one)

X ~ Bernoulli(p)

```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
```

Prior:

We first assume that the parameter "p" follows beta(3,3) distribution function

```python
a = 3 
b = 3
rv = beta(a, b)    #The prior distribution of the probability parameter of being one
```

```python
#Plotting the distribution of the probability parameter
p = np.linspace(0, 1, 101)
prior = rv.pdf(p)
plt.plot(p, prior, color = 'red')
plt.xlabel('p')
plt.ylabel('Probability')
```

Observed Data:


```python
obsrv = np.array([0 , 0 , 0 , 1 , 1 , 0, 1 , 1 , 1 , 1]) #Observed value of the random variable
```

Posterior:

We correct our assumption about the parameter using the observed data

```python
posterior = np.array([])
for i in prior:
    u = i
    for j in obsrv:
        u = u * pow(i, j) * pow(1-i, 1-j)    
    posterior = np.append(posterior, u) 
    
posterior = posterior/(sum(posterior) * 0.01)   
```

```python
plt.plot(p , posterior)
plt.xlabel('p')
plt.ylabel('Probability')
```
